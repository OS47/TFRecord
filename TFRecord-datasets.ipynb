### Functions to convert each feature type

```python
def _bytes_feature(value):
    if isinstance(value, type(tf.constant(0))):
        value = value.numpy() # BytesList won't unpack a string from an EagerTensor.
        return tf.train.Feature(bytes_list=tf.train.BytesList(value=[value]))

def _float_feature(value):
    return tf.train.Feature(float_list=tf.train.FloatList(value=[value]))

def _int64_feature(value):
    return tf.train.Feature(int64_list=tf.train.Int64List(value=[value]))
```

### Create a Features message using tf.train.Example. ( one training example)


```python
def serialize_example(image, label, image_shape):
    feature = {
        'image': _bytes_feature(image),
        'label': _int64_feature(label),
        'height': _int64_feature(image_shape[0]),
        'width': _int64_feature(image_shape[1]),
        'depth': _int64_feature(image_shape[2]),
    }
    example_proto = tf.train.Example(features=tf.train.Features(feature=feature))
    return example_proto.SerializeToString()
```

### Create TFrecord file

##### this code create bigger size of images cuz convert data to array then byte
* https://stackoverflow.com/questions/51292318/tfrecords-occupy-more-space-than-original-jpeg-images

```python
tfrecord_dir = 'data.tfrecords'
with tf.io.TFRecordWriter(tfrecord_dir) as writer:
    for image_path, label in zip(image_paths, labels):
        #load image
        img = tf.keras.preprocessing.image.load_img(image_path) 
        # convert to numpy
        img_array = tf.keras.preprocessing.image.img_to_array(img)
        # convert to byte data type    
        img_bytes = tf.io.serialize_tensor(img_array)
        #grap image shape [h,w,c]
        image_shape = img_array.shape
        
        example = serialize_example(img_bytes, label, image_shape)
        writer.write(example)
```

### Reading TFrecord file

```python
raw_image_dataset = tf.data.TFRecordDataset(tfrecord_dir)
```

```python
def read_tfrecord(serialized_example):
    feature_description = {
        'image': tf.io.FixedLenFeature((), tf.string),
        'label': tf.io.FixedLenFeature((), tf.int64),
        'height': tf.io.FixedLenFeature((), tf.int64),
        'width': tf.io.FixedLenFeature((), tf.int64),
        'depth': tf.io.FixedLenFeature((), tf.int64),
        
  }
    example = tf.io.parse_single_example(serialized_example, feature_description)
    
    image = tf.io.parse_tensor(example['image'], out_type = float)
    label = example['label']
    height = example['height']
    width = example['width']
    depth = example['depth']
  
    return image, label, height, width,depth
```

```python
parsed_dataset = raw_image_dataset.map(read_tfrecord)
for data in parsed_dataset.take(2):
    print(data)
```

```python
for image, label, height, width,depth in parsed_dataset.take(1):
    image = image.numpy() 
    label = label.numpy() 
    height =  height.numpy()
    width = width.numpy()
    depth = depth.numpy()
```
